```python
import pandas as pd
```

```python
# load datasets
resume_df = pd.read_csv("../data/raw_data/Resume.csv")
email_df = pd.read_csv("../data/raw_data/emails.csv")
company_doc_df = pd.read_csv("../data/raw_data/company-document-text.csv")

# look at the raw data
print("Resume DataFrame:")
display(resume_df.head())
print("Email DataFrame:")
display(email_df.head())
print("Company Document DataFrame:")
display(company_doc_df.head())
```

```python
# rename relevant fields to "content"
resume_df = resume_df.rename(columns={"Resume_str": "content"})
email_df = email_df.rename(columns={"text": "content"})
company_doc_df = company_doc_df.rename(columns={"text": "content"})

# drop all other fields
resume_df = resume_df[["content"]]
email_df = email_df[["content"]]
company_doc_df = company_doc_df[["content"]]

# apply labels 
resume_df["label"] = 1
email_df["label"] = 0
company_doc_df["label"] = 0

```

```python
# look at data again
print("Resume DataFrame:")
display(resume_df.head())
print("Email DataFrame:")
display(email_df.head())
print("Company Document DataFrame:")
display(company_doc_df.head())
```

```python
df = pd.concat([resume_df, email_df, company_doc_df], ignore_index=True)

df.to_csv("../data/clean_data/cleaned_data.csv", index=False)

display(df.head())
print(df.dtypes)

# data processing complete, ready for model training
```
